```python
import matplotlib.pyplot as plt
import json
%matplotlib inline 

with open("C:\\Users\zconn\Downloads\pluto.geojson", "r") as f:
    data = f.read()
    d = json.loads(data)
    for feature in d["features"]:
        pts = feature["geometry"]["coordinates"][0][0]
        x = [p[0] for p in pts]
        y = [p[1] for p in pts]
        plt.plot(x, y)  
    plt.show()
```
